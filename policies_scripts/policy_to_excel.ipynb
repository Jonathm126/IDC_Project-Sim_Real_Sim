```python
import json
from pathlib import Path
import pandas as pd
```

```python
policy_root = Path("/home/jonathan/Github/IDC_Project-Sim_Real_Sim/policies/act/so101_car_pick_and_place-bbox").resolve()
assert policy_root.exists()
```

```python
# Find configs
configs = sorted(policy_root.rglob("last/**/train_config*.json"))
if not configs:
    raise RuntimeError("No train_config*.json found under last/**/")

rows = []

for cfg_path in configs:
    with open(cfg_path, "r") as f:
        cfg = json.load(f)

    policy = cfg["policy"]
    inputs = policy["input_features"]

    run_name = policy.get("repo_id") or cfg_path.parents[2].name

    feature_desc = [
        f"{name}:{spec.get('shape')}"
        for name, spec in inputs.items()
    ]

    rows.append({
        "Run": run_name,
        "Num Input Features": len(inputs),
        "Input Features (name:shape)": "; ".join(feature_desc),
        "Action Shape": policy["output_features"]["action"]["shape"],
        "Model Dim": policy["dim_model"],
        "Encoder Layers": policy["n_encoder_layers"],
        "Decoder Layers": policy["n_decoder_layers"],
        "Use VAE": policy["use_vae"],
        "Latent Dim": policy["latent_dim"],
        "KL Weight": policy["kl_weight"],
        "Chunk Size": policy["chunk_size"],
        "N Action Steps": policy["n_action_steps"],
        "Training Steps": cfg["steps"],
        "Temporal Ensemble": policy.get("temporal_ensemble_coeff") is not None,
    })

df = pd.DataFrame(rows)
df
```
