```python
import torch
import sys
from pathlib import Path
from ultralytics import YOLO
!yolo settings sync=False;

# paths
sys.path.append(str(Path().resolve().parent))
from paths import REPO_ROOT, DATASETS_DIR, POLICIES_DIR

# set up env secrets
from dotenv import load_dotenv
load_dotenv(REPO_ROOT / ".env", override=True)
torch.cuda.empty_cache()
```

```python
PUSH_TO_HUB     = False
REPO_NAME       = 'so101_car_pick_and_place'
EXPERIMENT_NAME = 'v0'
DATASET_PATH    = DATASETS_DIR / REPO_NAME
BBOX_DATA_PATH  = DATASET_PATH / 'bboxes' / 'data.yaml'
YOLO_SAVE_PATH  = POLICIES_DIR / 'yolo'
assert(BBOX_DATA_PATH.exists())
```

```python
# Load a model
res = 640 
model_size = 's'
model = YOLO(f'yolov8{model_size}-obb.pt')

results = model.train(
    data         = BBOX_DATA_PATH,
    epochs       = 200,
    batch        = 16,
    imgsz        = res,
    device       = 'cuda',
    workers      = 8,
    plots        = True,
    project      = YOLO_SAVE_PATH,
    name         = f'yolov8{model_size}-obb_{EXPERIMENT_NAME}',
    exist_ok     = True,
    auto_augment = True,
    augment      = True,
    fraction     = 1.0
    )
```

```python
results = model.val(data=BBOX_DATA_PATH)
```
