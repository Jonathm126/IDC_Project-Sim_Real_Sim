```python
from safetensors import safe_open
from safetensors.torch import save_file
```

```python
# Load your safetensors file
model_path = "../policies/act/so101_pick_pen-bbox/bbox_v0/checkpoints/last/pretrained_model/policy_preprocessor_step_3_normalizer_processor.safetensors"

keys_to_change = [
    "observation.state.max",
    "observation.state.mean",
    "observation.state.min",
    "observation.state.std",
]

new_tensors = {}

with safe_open(model_path, framework="pt") as f:
    for key in f.keys():
        tensor = f.get_tensor(key)

        if key in keys_to_change:
            # keep first 6 values, remove last 6
            tensor = tensor[:6]

        new_tensors[key] = tensor
```

```python
save_file(new_tensors, model_path)
```

```python
with safe_open(model_path, framework="pt") as f:
    for key in keys_to_change:
        print(key, f.get_tensor(key).shape)
```
