```python
dfmeta = pd.read_parquet(ds.root.__str__() + '/meta/episodes/chunk-000/file-000.parquet')
dfmeta.head(10)
```

```python
dfdata = pd.read_parquet(ds.root.__str__() +'/data/chunk-000/file-000.parquet')
dfdata.head(100)
```

```python
arr = np.stack(dfdata['observation.environment_state'].values) 
maxs = np.array([640.0, 480.0, 360.0])
mins = np.array([0.0, 0.0, 0.0])
arr_norm = (arr - mins) / (maxs - mins)
dfdata_norm = dfdata.copy(deep=True)
dfdata_norm['observation.environment_state'] = list(arr_norm.astype(np.float32))
```
